```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import scipy
from scipy import signal
```

# Generation of unit impulse signal

```python
x = []
t = []

for i in range(-5,6): 
	t.append(i)
	if i == 0: 
		x.append(1)
	else: 
		x.append(0)
```

```python
print(x)
print(t)
```

```python
plt.figure(figsize= (18,8))

style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x,linewidth = 3, label = 'Unit impulse function')
plt.ylim(-0.01, 1)
plt.xlim(-5,5)
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

## Another method of generating impulse
We can just use the signal library to generate.

```python
Impulse = signal.unit_impulse(10,'mid')
```

```python
plt.figure(figsize = (18,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(np.arange(-5,5), Impulse, linewidth = 3, label = 'Unit impulse function')
plt.ylim(-0.01,1)
plt.xlabel('time', fontsize =15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

# Time shifted Impulse (Time delay)

```python
shifted_impulse = signal.unit_impulse(7,2) # 7 is the number of time points. when t =2 is when we will get the signal
shifted_impulse
```

```python
plt.figure(figsize = (18,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(shifted_impulse, linewidth = 3, label = 'Shifted impulse function (Time delay)')
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

## Time Shifted Impulse (Time advance)

```python
x = []
t = []

for i in range(-5,6): 
	t.append(i)
	if i == -2: 
		x.append(1)
	else: 
		x.append(0)
```

```python
plt.figure(figsize = (18,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x, linewidth = 3, label = 'Shifted impulse function (Time advance)')
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

## Generation of unit step signal

```python
x = []
t = []

for i in range(-5,10):
	t.append(i)
	if i >= 0: 
		x.append(1)
	else: 
		x.append(0)
```

```python
print(x)
print(t)
```

```python
plt.figure(figsize = (18,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x, linewidth = 3, label = 'Unit step function')
plt.ylim(0,1.5)
plt.xlim(-5,9)
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

## Generation of Ramp Signal

```python
x = []
t = []
for i in range(0,10): 
	t.append(i)
	x.append(i)
```

```python
print(x)
print(t)
```

```python
plt.figure(figsize = (18,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x, linewidth = 3, label = 'Ramp function')
plt.ylim(0,len(x))
plt.xlim(0,len(t)-1)
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

## Generation of Sine wave

```python
t = np.linspace(0,10,100)
amp = 5 
f = 50
x = amp * np.sin(2 * np.pi * f * t)
```

```python
plt.figure(figsize = (15,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x, linewidth = 3, label = 'SineWave')
plt.ylim(-5,5)
plt.xlim(0,10)
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```

## Generation of exponential signal

```python
t = np.linspace(0,10,100)
amp = 1
x = amp * np.exp(-t)
```

```python
plt.figure(figsize = (15,8))
style.use('dark_background')
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.plot(t,x, linewidth = 3, label = 'Exponential Signal')
plt.ylim(0,1)
plt.xlim(0,5)
plt.xlabel('time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.legend(fontsize = 20)
plt.show()
```
