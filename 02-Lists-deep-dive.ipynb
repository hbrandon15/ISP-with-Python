# Lists

Lists are more flexible than arrays because it can store different data types.

```python
first_list = [1,2,3]

second_list = [7, 'hello', 4.5]

print(first_list)
print(second_list)
```

Similar to arrays, we can index and slice lists. 

```python
first_list[0] # print the first element
```

```python
second_list[1:] # print the second element till the end of the list
```

### Adding elements to a list

```python
first_list  = first_list + [4]
print(first_list)
```

### Applying methods on a list

```python
list = [1,2,3]
```

```python
list.append(5) # add the number 5 to the end of the list
print(list)
```

```python
list.pop() # remove the last element of the list
print(list)
```

```python
# you can also pop a specific index
list.pop(1) # removing the element at index 1 
print(list)
```

```python
list.reverse() # reverse the list order
print(list)
```

```python
list.sort()
print(f"Sorted list: {list}")
```

### Nested list

```python
my_nested_list = [1,2,3,4,[6,7,8]]
print(my_nested_list)
```

Other methods:

```python
# Lets make three lists

list_1 = [1,2,3]
list_2 = [4,5,6]
list_3 = [7,8,9]

# make a list of list to form a matrix

matrix = [list_1, list_2, list_3]
print(matrix)

print(f"\nIt is important to NOTE, the length of this matrix is: {len(matrix)}")
```

### Indexing in Nested List

```python
new_list = [1,2,3,4,[11,12,13]]
print(new_list)
```

```python
# lets print the 4th element in our new list
print(new_list[4])
```

```python
# now lets get the 4th element, and the first index of the nested list
print(new_list[4][0])
```
