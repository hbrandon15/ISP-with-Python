```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,10) # give me 10 numbers between 0-5
y = 2 * x # double the values in x
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y)
```

### Second plot on the top of the first plot

```python
plt.plot(x,y)
plt.plot(y,x)
```

### Display two plots separately

```python
plt.plot(x,y)
plt.show()
plt.plot(y,x)
```

### Change color of the plot

```python
y = x**2
plt.plot(x,y,'g') # change the color of the plot to green
```

### Change the size of the plot

```python
y = x ** 2
plt.figure(figsize= (10,8)) # Set the plot size to a 10x8
plt.plot(x,y, 'r') 
```

### Increase the width of the plot line

```python
y = x ** 2
plt.figure(figsize= (4,6)) # Set the plot size to a 10x8
plt.plot(x,y, 'r', linewidth = 3) 
```

### Label X-axis & Y-axis

```python
y = x ** 2
plt.figure(figsize= (10,8)) # Set the plot size to a 10x8
plt.plot(x,y, 'r')
plt.xlabel('Time', fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
```

### Making plot more informative by legend and title

```python
y = x ** 2
plt.figure(figsize= (10,8)) # Set the plot size to a 10x8
plt.plot(x,y, 'r', label = 'y = x **2')
plt.xlabel('X', fontsize = 15)
plt.ylabel('Y', fontsize = 15)
plt.legend(fontsize = 20)
plt.title('Visualization of y = x ** 2', fontsize = 25)
plt.show() 
```
