### Generating signals for subplotting

```python
x = np.linspace(0,5,10)
y1 = x ** 2
y2 = x ** 3
```

```python
plt.subplot(2,1,1) # rows, columns, position
plt.plot(x,y1) # plotting on position  1
plt.subplot(2,1,2)
plt.plot(x,y2) # plotting on position 2
plt.tight_layout()
```

### Styling with matplotlib rcParams

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
x = np.linspace(0,5,10)
y1 = x ** 2
y2 = x ** 3
```

```python
plt.figure(figsize = (10,6))
plt.subplot(2,1,1) # (rows, columns, position)
plt.plot(x,y1,lw = 3, ls = '--')

plt.subplot(2,1,2)
plt.plot(x,y2,lw = 3, ls = '--')
plt.show()
```

### Using rcParams

```python
# rcParams allows you to apply global styles to plots
plt.figure(figsize = (10,6))
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.linestyle'] = '--'
plt.subplot(2,1,1) # (rows, columns, position)
plt.plot(x,y1)

plt.subplot(2,1,2)
plt.plot(x,y2)
plt.show()
```
